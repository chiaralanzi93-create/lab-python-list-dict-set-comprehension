````python
#4
def update_inventory(inventory, customer_orders):
    # Create a new inventory dictionary with updated quantities
    updated_inventory = {product: (inventory[product] - 1) for product in customer_orders if product in inventory}
    
    # Filter products with a quantity greater than zero
    updated_inventory = {product: qty for product, qty in updated_inventory.items() if qty > 0}

    # Include products that weren't ordered
    for product, qty in inventory.items():
        if product not in customer_orders:
            updated_inventory[product] = qty

    return updated_inventory

# Assuming inventory and customer_orders are defined before this
updated_inventory = update_inventory(inventory, customer_orders)
print("The updated inventory is:", updated_inventory)
````
